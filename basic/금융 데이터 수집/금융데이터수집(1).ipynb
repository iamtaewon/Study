```python
# 전체 종목의 펀더멘탈 지표(PER, PBR, EPS, BPS, DIV, DPS) 가져오기
# stock.get_market_fundamental_by_ticker(date=날짜, market=거래소)

# market의 기본값은 "ALL"

# DIV(배당수익률): (주가배당금/주가) * 100

# BPS(주당순자산가치=청산가치): (순자산)/(총발행주식수)

# PER(주가수익비율): (주가)/(주당순이익)

# EPS(주당순이익): (당기순이익)/(총발행주식수)

# PBR(주가순자산비율) = (주가)/(BPS) = PER*EPS / BPS
```

```python
df = stock.get_market_fundamental_by_ticker(date="20230327", market="KOSPI")
df.head()
```

```python
# 특정일자의 종목별 DIV/BPS/PER/EPS 조회
# get_market_fundamental_by_ticker (date="YYYYMMDD", market="거래소명")
```

```python
df = stock.get_market_fundamental_by_ticker(date="20230327")
df.head()
```

```python
# 일자별 개별종목 DIV/BPS/PER/EPS 조회
# get_market_fundamental_by_date(fromdate, todate, ticker, freq='d', name_display=False)

# freq: d(일), m(월), y(연도)
```

```python
#지정된 기간 동안의 일자별로 정렬된 (순매수) 거래량
df = stock.get_market_trading_value_by_date("20230101", "20230327", "005930")
print(df.tail())
```

```python
# 매수 혹은 매도 금액을 조회하고 싶다면 on 파라미터에 입력할 수 있습니다. 입력하지 않는다면 순매수 금액이 조회
df = stock.get_market_trading_value_by_date("20230101", "20230327", "005930", on='매도')
print(df.tail())
```

```python
# 매수 혹은 매도 금액을 조회하고 싶다면 on 파라미터에 입력할 수 있습니다. 입력하지 않는다면 순매수 금액이 조회
df = stock.get_market_trading_value_by_date("20230101", "20230327", "005930", on='매수')
print(df.tail())
```

```python
#일자별 시장의 거래량을 조회
df = stock.get_market_trading_volume_by_date("20230101", "20230327", "005930")
print(df.tail())
```

```python
# 매수 혹은 매도 거래량만을 조회하고 싶다면 on 파라미터에 입력 가능
# 입력하지 않는다면 순매수 수량이 조회
df = stock.get_market_trading_volume_by_date("20230101", "20230327", "005930", on='매도')
print(df.head())
```

```python
#지정된 기간 동안의 투자자별로 정렬된 (순매수) 거래대금
df = stock.get_market_trading_value_by_investor("20230101", "20230327", "005930")
print(df.tail())
```

```python
#세 번째 파라미터에 KOSPI, KOSDAQ, KONEX를 지정하면 각 시장의 거래대금을 조회
# ALL을 입력하면 세 개의 시장을 합친 통합 거래대금을 조회
df = stock.get_market_trading_value_by_investor("20230101", "20230327", "KOSPI")
print(df.tail())
```

```python
#시장의 거래대금을 조회할 때는 ETF, ETN ELW의 포함 여부를 지정 가능
#만약 값을 지정하지 않는다면 해당 데이터를 제외한 값을을 반환
# 옵션은 KOSPI, KOSDAQ, KONEX 조회할 때만 유효
df = stock.get_market_trading_value_by_investor("20230101", "20230327", "KOSPI", etf=True, etn=True, elw=True)
print(df.tail())
```

```python
import yfinance as yf

ticker = yf.Ticker('005930.KS')
 
ticker.history(
               interval='1d',
               start='2022-01-01',
               end='2022-03-24',
               actions=True,
               auto_adjust=True)
```

```python
# 분단위 데이터 조회. 조회할 수 있는 기간에 제한이 있음

df_minute = ticker.history(
               interval='1m',
               start='2023-03-10',
               end='2023-03-15',
               actions=True,
               auto_adjust=True)
df_minute
```
