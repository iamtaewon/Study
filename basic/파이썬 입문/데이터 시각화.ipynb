```python
# 한 화면에 여러 개의 그래프 그리기
# figure 함수를 통해 Figure 객체를 먼저 만든 후 
# add_subplot 메서드를 통해 그리려는 그래프 개수만큼 subplot을 만든다.
```

```python
# subplot의 개수는 add_subplot 메서드의 인자를 통해 조정 가능
# 만약 (2, 1, 1)은 2x1(행x열)의 subplot을 생성한다는 의미이고 세 번째 인자 1은 생성된 두 개의 subplot 중 첫 번째 subplot을 의미 
# 마찬가지로 (2, 1, 2)는 2x1 subplot에서 두 번째 subplot을 의미
```

```python
# 변수 명명 : ax1, ax2라는 변수를 통해 생성된 객체 (Axes 용어)
fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)
plt.show()
# Figure 안에 두 개의 subplot이 존재하는데, 행 방향으로 2개, 열 방향으로는 1개의 subplot이 위치함을 확인
```

```python
# 2x1 subplot
# add_subplot 메서드의 인자를 변경해 1x2 형태의 subplot을 만드는 경우
# 1x2 형태의 subplot이 생성되는 것을 확인

fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
plt.show()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

x = range(0, 100)
y = [a*a for a in x]

# bar 함수는 pyplot 모듈에 있는 함수로 막대그래프를 그리는 데 사용
ax1.plot(x,y)
ax2.bar(x,y)

plt.show()
```
